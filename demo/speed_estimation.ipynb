```python
# Импортируемые библиотеки
import cv2
from ultralytics import YOLO, solutions
```

```python
video_path = "../test_videos/road.mp4" # Путь до видео
output_video_path = "../output_videos/speed_estimation_output.mp4" # Путь до выходного видео
weights_path = "../weights/yolov10n.pt" # Путь до весов модели
```

```python
model = YOLO(weights_path)
names = model.model.names

cap = cv2.VideoCapture(video_path)
assert cap.isOpened(), "Ошибка при чтении видеофайла"
w, h, fps = (int(cap.get(x)) for x in (cv2.CAP_PROP_FRAME_WIDTH, cv2.CAP_PROP_FRAME_HEIGHT, cv2.CAP_PROP_FPS))

video_writer = cv2.VideoWriter(output_video_path, cv2.VideoWriter_fourcc(*"mp4v"), fps, (w, h))

line_pts = [(0, 360), (1280, 360)]

speed_obj = solutions.SpeedEstimator(
    reg_pts=line_pts,
    names=names,
    view_img=True,
)

while cap.isOpened():
    success, im0 = cap.read()
    if not success:
        print("Кадр видео пустой или обработка видео прошла успешно.")
        break

    tracks = model.track(im0, persist=True)

    im0 = speed_obj.estimate_speed(im0, tracks)
    video_writer.write(im0)

    # Показ окна с кадром
    cv2.imshow("Speed Estimation", im0)

    # Обработка нажатия клавиши
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
video_writer.release()
cv2.destroyAllWindows()
```
