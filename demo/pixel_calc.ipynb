```python
import cv2

from ultralytics import YOLO, solutions
```

```python
w, h, fps = (int(cap.get(x)) for x in (cv2.CAP_PROP_FRAME_WIDTH, cv2.CAP_PROP_FRAME_HEIGHT, cv2.CAP_PROP_FPS))

model = YOLO("../weights/yolov10n.pt") # Выберите нужные веса
cap = cv2.VideoCapture("../test_videos/test.mp4") # Путь до видео
video_writer = cv2.VideoWriter("../output_videos/pixel_calculation_output.mp4", cv2.VideoWriter_fourcc(*"mp4v"), fps, (w, h)) # Путь до выходного видео
```

```python
names = model.model.names

assert cap.isOpened(), "Ошибка в чтении файла"

# Инициализация просчета дистанции
dist_obj = solutions.DistanceCalculation(names=names, view_img=True)

while cap.isOpened():
    success, im0 = cap.read()
    if not success:
        print("Кадр видео пустой или обработка видео прошла успешно.")
        break

    tracks = model.track(im0, persist=True, show=False)
    im0 = dist_obj.start_process(im0, tracks)
    video_writer.write(im0)
    
    
    cv2.imshow("Pixel calculation", im0)

    # Завершение по нажатию клавиши 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
video_writer.release()
cv2.destroyAllWindows()
```
